## (b)

The $N$-period binomial model is a complete model, that is, every contingent claim can be hedged by some trading strategy. The replicating portfolio will have an initial stage $(\phi_0, \psi_0)$ and two later stages $(\phi_1, \psi_1)$ depending on whether the price of the stock at time $t = 1$ has gone up ($S_1 = uS_0$) or down ($S_1 = dS_0$). 

We compute first the risk neutral probability:

$$
q = \frac{e^r - d}{u - d} = \frac{\tfrac{3}{4} - \tfrac{1}{2}}{\tfrac{5}{4} - \tfrac{1}{2}} = \tfrac{1}{3}.

The portfolio values are given by the following formulas and must be applied for the two possible values of the stock at time $t = 1$:

$$\phi_2(S_0, S_1) = \frac{V_2(S_0, S_1, uS_1) - V_2(S_0, S_1, dS_1)}{S_1(u-d)}$$

$$\psi_2 (S_0, S_1) = \frac{uV_2(S_0, S_1, dS_1) - d V_2(S_0, S_1, uS_1)}{e^{2r}(u-d)}$$

We compute the replicating portfolio $(\phi_2, \psi_2)$ for $S_1 = uS_0 = 11.25$. 

$$\phi_2(S_0 = 9, S_1 = 11.25) = \frac{5 - 1.625}{11.25\cdot 0.75} = 0.4$$

$$\psi_2 (S_0 = 9, S_1 = 11.25) = \frac{1.25\cdot  1.625 - 0.5\cdot 5}{0.5625\cdot 0.75} = -\frac{10}{9} \approx -1.11$$




We can also see that the portfolio replicates the payoff for $S_2 \in \{14.0625, 5.625\}$  given that $S_1 = uS_0 = 11.25$:

$$V_2 = \phi_2S_2 + \psi_2 e^{2r},\quad S_2 \in \{14.0625, 5.625\}.$$

$$V_2(S_2 = uS_1) = 0.4 \cdot 14.0625  -\frac{10}{9} \cdot 0.5625 = 5$$
$$V_2(S_2 = dS_1) = 0.4 \cdot 5.625 -\frac{10}{9} \cdot 0.5625 = 1.625$$

In both cases, for $S_1 = 11.25$, the replicating portfolio and the payoff agree.

With the self-financing condition we can also check that the values of the portfolio at time $t = 1$ also coincide.

$$V_n = \phi_nS_n + \psi_n e^{nr} = \phi_{n+1}S_n + \psi_{n+1}e^{nr}.$$
$$V_1(S_0 = 9, S_1 = 11.25) = \phi_2 S_1 + \psi_2 e^{r} = 0.4 \cdot 11.25 - \frac{10}{9}\cdot 0.75 = \frac{11}{3} \approx 3.67.$$

We compute the replicating portfolio $(\phi_2, \psi_2)$ for $S_1 = dS_0 = 4.5$. 



$$\phi_2(S_0 = 9, S_1 = 4.5) = \frac{0.5 - 0}{4.5\cdot 0.75} = \frac{4}{27} \approx 0.15$$

$$\psi_2 (S_0 = 9, S_1 = 4.5) = \frac{1.25\cdot  0 - 0.5\cdot 0.5}{0.5625\cdot 0.75} = -\frac{16}{27} \approx -0.59$$


We can also see that the portfolio replicates the payoff for $S_2 \in \{5.625, 2.25\}$ given that $S_1 = dS_0 = 4.5$:

$$V_2 = \phi_2S_2 + \psi_2 e^{2r},\quad S_2 \in \{5.625, 2.25\}.$$

$$V_2(S_2 = uS_1) = \frac{4}{27} \cdot 5.625  -\frac{16}{27} \cdot 0.5625 = 0.5$$
$$V_2(S_2 = dS_1) = \frac{4}{27} \cdot 2.25 -\frac{16}{27} \cdot 0.5625 = 0$$

In both cases, for $S_1 = 4.5$, the replicating portfolio and the payoff agree.

Finally, we can compute the portfolio couple for time $t = 1$:

$$\phi_1(S_0 = 9) = \frac{11/3 - 2/9}{9\cdot 3/4} = \frac{124}{243} \approx 0.51$$
$$ \psi_1(S_0 = 9) = \frac{5/4 \cdot 2/9 - 1/2 \cdot 11/3}{3/4\cdot3/4} = -\frac{224}{81}$$

$$V_0 = \phi_1 S_0 + \psi_1 = \frac{124}{243} \cdot 9 -\frac{224}{81}  = \frac{148}{81}\approx 1.83$$

$$\mathbb{E}\left[\left(Se^{\mu + \tau Z} - K\right)^{+}\right] = Se^{\mu + \frac{\tau^2}{2}} \Phi(\tau - z_0) - K \Phi(-z_0).$$
\textit{Using this, derive the (asymptotic) Black-Scholes formula from Example 4.19.}


\begin{align*}
	\mathbb{E}\left[\left(Se^{\mu + \tau Z} - K\right)^{+}\right] &= \mathbb{E}\left[Se^{\mu + \tau Z}\mathbb{1}_{\{Se^{\mu + \tau Z} > K \}}\right] - \mathbb{E}\left[ K\mathbb{1}_{\{Se^{\mu + \tau Z} > K \}}\right]\\
	&= \mathbb{E}\left[Se^{\mu + \tau Z}\mathbb{1}_{\{Z > \tau^{-1}\left(\log(\frac{K}{S}) - \mu\right) \}}\right] - K\, \mathbb{E}\left[\mathbb{1}_{\{Z > \tau^{-1}\left(\log(\frac{K}{S}) - \mu\right) \}}\right]\\
	&= \underbrace{\mathbb{E}\left[Se^{\mu + \tau Z}\mathbb{1}_{\{Z > z_0\}}\right]}_{I} - \underbrace{ K\,\mathbb{E}\left[\mathbb{1}_{\{Z > z_0\}}\right]}_{II}
\end{align*}

We compute $I$ making use of the normal's pdf symmetry:
\begin{align*}
K\,\mathbb{E}\left[\mathbb{1}_{\{Z > z_0\}}\right] =  K\mathbb{P}\left(Z > z_0\right) = K\mathbb{P}\left(Z < -z_0\right) = K\Phi(-z_0)
\end{align*}

We compute $II$ evaluating the expression and with the change of variables $u = z - \tau$: 
\begin{align*}
\mathbb{E}\left[Se^{\mu + \tau Z}\mathbb{1}_{\{Z > z_0\}}\right] &= \int_{z_0}^{+\infty} Se^{\mu + \tau Z} \frac{1}{\sqrt{2\pi}}e^{-\frac{z^2}{2}}\,dz = Se^{\mu} \int_{z_0}^{+\infty}\frac{1}{\sqrt{2\pi}}  e^{\tau Z - \frac{z^2}{2}}\,dz \\ &=  Se^{\mu} \int_{z_0}^{+\infty}\frac{1}{\sqrt{2\pi}}  e^{-\frac{1}{2}(z - \tau)^2 + \frac{\tau^2}{2}}\,dz = Se^{\mu + \frac{\tau^2}{2}} \int_{z_0 - \tau}^{+\infty}\frac{1}{\sqrt{2\pi}}  e^{-\frac{1}{2}u^2}\,du\\
&=Se^{\mu + \frac{\tau^2}{2}} \mathbb{P}\left(Z > (z_0 - \tau)\right) = Se^{\mu + \frac{\tau^2}{2}} \mathbb{P}\left(Z < (\tau - z_0 )\right) = Se^{\mu + \frac{\tau^2}{2}} \Phi(\tau - z_0).
\end{align*}

Putting $I$ and $II$ together we get the desired result:

$$\mathbb{E}\left[\left(Se^{\mu + \tau Z} - K\right)^{+}\right] = Se^{\mu + \frac{\tau^2}{2}} \Phi(\tau - z_0) - K \Phi(-z_0).$$

To compute the fair price of a European call option from Example 4.19, we obtain the expression of $\bar{S}_T$ as a function of a normal random variable $Z \sim N(0, 1)$, the mean $a$ and the square root of the variance $b$ since we now that $ \log\left(\frac{\bar{S}_T}{S_0}\right) \sim N((r-\frac{\sigma^2}{2})T, \sigma^2 T)$:

\begin{align*}
	&\log\left(\frac{\bar{S}_T}{S_0}\right) =  a + Z b \Rightarrow \bar{S}_T =  S_0 e^{(r-\frac{\sigma^2}{2})T + \sigma \sqrt{T} Z}
\end{align*}
Now based on the previous exercise, define $\mu := (r-\frac{\sigma^2}{2})T$, $\tau:= \sigma \sqrt{T}$ and $z_0 := \frac{\log(K/S_0) + \frac{\sigma^2}{2}T - rT}{\sigma \sqrt{T}}$. The expected value of the payoff function:
\begin{align*}& \mathbb{E}\left[e^{-rT}\left(\bar{S}_T -K \right)^{+}\right] =\mathbb{E}\left[e^{-rT}\left(S_0 e^{(r-\frac{\sigma^2}{2})T + \sigma \sqrt{T} Z} -K \right)^{+}\right] = \mathbb{E}\left[e^{-rT}\left(S_0 e^{\mu + \tau Z} -K \right)^{+}\right]\\ &= e^{-rT} \left[S_0 e^{\mu + \frac{\tau^2}{2}} \Phi(\tau - z_0) - K \Phi(-z_0)\right]\\  &= S_0 e^{(r-\frac{\sigma^2}{2})T + \frac{\sigma^2 T}{2} - rT} \Phi\left(\sigma \sqrt{T} - \frac{\log(K/S_0) + \frac{\sigma^2}{2}T - rT}{\sigma \sqrt{T}}\right) - e^{-rT}K \Phi\left(-\frac{\log(K/S_0) + \frac{\sigma^2}{2}T - rT}{\sigma \sqrt{T}}\right)\\

```python
import numpy as np
```

```python
def put_option(S_0, r1, u, d, T, K):
    # This function simulates all possible stock trajectories in a T-period binomal model and 
    # determines the corresponding European put option fair prices corresponding to the payoffs
    
    # S_0 is the initial stock price
    # r_1 is the daily interest rate used as 1+r1 multiplier
    # u is the amount by which the stock goes up
    # d is the amount by which the stock goes down
    # T is the maturity time
    # K is the strike price of the put option
    # stockprice is 2 dimensional numpy array containing all possible stock values throughout the time steps up until time T

    q = (1+r1-d)/(u-d) #risk neutral probability for up
    
    stockprice = np.full((T+1,T+1),np.nan)
    stockprice[0,0] = S_0
    for j in range(1,T+1):
        for i in range(0,j):
            stockprice[i,j] = stockprice[i,j-1]*u
            stockprice[i+1,j] = stockprice[i,j-1]*d
    print("Stock prices:")
    print(np.around(stockprice, 2)) #just for visualisation

    putoption = np.full((T+1,T+1),np.nan) 
    putoption[:,-1] = np.where(K-stockprice[:,-1]<0,0,K-stockprice[:,-1]) # V_0
    for j in range(T-1,-1,-1):
        for i in range(0,j+1):
            putoption[i,j] = (q*putoption[i,j+1]+(1-q)*putoption[i+1,j+1])/(1+r1) # Explicit expected value
            # print(i, j)
            # print(putoption[i,j])
    print("Put option prices:")
    print(np.around(putoption, 2)) #just for visualisation
    return stockprice, putoption
```

```python
S_0 = 1000
r1 = 0.02/365
u_values = [1.02, 1.03, 1.05, 1.101]
u = 1.03
d = 0.99
T = 7
K = 990
```

```python
import matplotlib.pyplot as plt

d_values = [0.99, 0.95]
u_values = [1.03, 1.1, 1.2, 1.3]

for d in d_values:
    for u in u_values:
        stockprice, putoption = put_option(S_0, r1, u, d, T, K)
        plt.imshow(putoption, cmap="viridis", interpolation="nearest")
        plt.colorbar(label="Put Option Price")
        plt.title(f"Binomial Tree Heatmap of Put Prices\n(u = {u}, d = {d})")
        plt.xlabel("Time Step")
        plt.ylabel("Time Step")
        plt.show()
```

**In summary:**
- Higher `u` → lower put price  
- Lower `d` → higher put price
