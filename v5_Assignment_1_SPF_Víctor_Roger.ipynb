The portfolio values are given by the following formulas and must be applied for the two possible values of the stock at time $t = 1$:

$$\phi_2(S_0, S_1) = \frac{V_2(S_0, S_1, uS_1) - V_2(S_0, S_1, dS_1)}{S_1(u-d)}$$

$$\psi_2 (S_0, S_1) = \frac{uV_2(S_0, S_1, dS_1) - d V_2(S_0, S_1, uS_1)}{e^{2r}(u-d)}$$

We compute the replicating portfolio $(\phi_2, \psi_2)$ for $S_1 = uS_0 = 11.25$.

$$\phi_2(S_0 = 9, S_1 = 11.25) = \frac{5 - 1.625}{11.25\cdot 0.75} = 0.4$$

$$V_2 = \phi_2S_2 + \psi_2 e^{2r},\quad S_2 \in \{14.0625, 5.625\}.$$

$$V_2(S_2 = uS_1) = 0.4 \cdot 14.0625  -\frac{10}{9} \cdot 0.5625 = 5$$
$$V_2(S_2 = dS_1) = 0.4 \cdot 5.625 -\frac{10}{9} \cdot 0.5625 = 1.625$$

In both cases, for $S_1 = 11.25$, the replicating portfolio and the payoff agree.

With the self-financing condition we can also check that the values of the portfolio at time $t = 1$ also coincide.

$$V_n = \phi_nS_n + \psi_n e^{nr} = \phi_{n+1}S_n + \psi_{n+1}e^{nr}.$$
$$V_1(S_0 = 9, S_1 = 11.25) = \phi_2 S_1 + \psi_2 e^{r} = 0.4 \cdot 11.25 - \frac{10}{9}\cdot 0.75 = \frac{11}{3} \approx 3.67.$$

## (d)

If $P > V_0$, then an explicit arbitrage strategy is:

- At $t = 0$:
  Sell the contract for $P$. Use $V_0$ to set up the replicating portfolio consisting of $
  \phi_0$ shares of stock and $\psi_0$ in the bank account.
  We are left with $P - V_0 > 0$  funds in the bank.

- At $t = 1$:
  Rebalance the replicating portfolio to the corresponding $(\phi_1, \psi_1)$ values
  depending on whether $S_1 = S_0u$ or $S_1 = S_0d$.

- At $t = 2$:
  The replicating portfolio pays exactly the contract payoff, so the short position
  in the contract and the portfolio cancel perfectly.  
  The initial surplus $P - V_0$ has grown risk-free to $(P - V_0)e^{2r} > 0$.

Hence, starting from zero initial wealth, the strategy yields a sure positive profit, i.e., an arbitrage strategy.

First, observe that, for any $\mathbb{P} \sim \mathbb{Q}$, $\mathbb{P}$ is free of arbitrage $\Leftrightarrow$ $\mathbb{Q}$ is free of arbitrage. Let’s show this:

$$
\begin{cases}
\mathbb{P}(V_n > 0) > 0 \Leftrightarrow \mathbb{Q}(V_n \neq 0) >  0 \Leftrightarrow \mathbb{Q}(V_n > 0) > 0 \\
\mathbb{P}(V_0 = 0) = 1 \Leftrightarrow \mathbb{P}(V_0 \neq 0) = 0 \Leftrightarrow \mathbb{Q}(V_0 \neq 0) = 0 \Rightarrow \mathbb{Q}(V_0 = 0) = 1 \\
\mathbb{P}(V_n \geq 0) = 1 \Leftrightarrow \mathbb{P}(V_n < 0) = 0 \Leftrightarrow \mathbb{Q}(V_n < 0) = 0 \Rightarrow \mathbb{Q}(V_n \geq 0) = 1
\end{cases}
$$

We will now prove that $d<e^r<u \implies $ No arbitrage under $\mathbb{P}\sim\mathbb{Q}$.

Let $d \in \mathbb{R}^u$. We will show that there’s no arbitrage in $\mathbb{Q}$:

Let $(V_n)_n$ be such that $\mathbb{P}(V_0 = 0) = 1$ and $\mathbb{P}(V_n \ge 0) = 1$. We will see that $\mathbb{P}(V_n > 0) = 0$:

$$
\mathbb{E}_{\mathbb{Q}}[V_N] = \mathbb{E}_{\mathbb{Q}}[\mathbb{E}_{\mathbb{Q}}[V_N | \mathcal{F}_0]] = \mathbb{E}_{\mathbb{Q}}[\mathbb{E}_{\mathbb{Q}}[e^{r N}\tilde{V_N} | \mathcal{F}_0]]
= \mathbb{E}_{\mathbb{Q}}\left[e^{r N} \tilde{V_0}\right] = e^{r N} V_0 = 0,
$$
where we used proposition 4.4, which states that the discounted value of the binomial model is a martingale under $\mathbb{Q}$ if $d<e^r<u$.

Observe that, since $V_N$ is non-negative a.s., its expected value must be non-negative. Since the expected value is equal to $0$, we deduce that $V_n \equiv 0$ a.s., i.e.
$$
\mathbb{P}(V_n > 0) = 0.
$$

Now we will use proposition 2.7 to prove that, if $\mathbb{P}\sim\mathbb{Q}$ is free of arbitrage, then $d<e^r<u$. We will show that, if $d<e^r<u$ does not hold, then there exists an arbitrage opportunity.

By proposition 2.7, we know that, if $d<e^r<u$ doesn't hold, then there exists a portfolio $(\phi_n,\psi_n)$ with value $V_n$, such that:

$$
\begin{cases}
\mathbb{P}(V_1 > 0) > 0  \\
\mathbb{P}(V_0 = 0) = 1 \\
\mathbb{P}(V_1 \geq 0) = 1.
\end{cases}
$$
Using said portfolio, we can construct a N-period replicating portfolio $V'_n$ which, until $t=1$ replicates $V_n$ and allocates all its value to the risk-free asset afterwards. This way, $V'_n = e^{(n-1)\, r}\,V_1$ for all $n = 1,\ldots,N$. Then:
$$
\begin{cases}
\mathbb{P}(V'_N > 0)  =  \mathbb{P}(e^{(N-1)\, r}V_1 > 0) = \mathbb{P}(V_1 > 0) > 0\\
\mathbb{P}(V'_0 = 0) = \mathbb{P}(V_0 = 0) = 1 \\
\mathbb{P}(V'_N \geq 0) = \mathbb{P}(e^{(N-1)\, r}V_1 \geq 0)= \mathbb{P}(V_1 \geq 0) = 1.
\end{cases}
$$
Therefore, we have shown that, if $d<e^r<u$ doesn't hold, then there exists arbitrage under $\mathbb{P}$, which concludes the proof.

```python
import numpy as np
```

```python
def put_option(S_0, r1, u, d, T, K):
    # This function simulates all possible stock trajectories in a T-period binomal model and
    # determines the corresponding European put option fair prices corresponding to the payoffs

    # S_0 is the initial stock price
    # r_1 is the daily interest rate used as 1+r1 multiplier
    # u is the amount by which the stock goes up
    # d is the amount by which the stock goes down
    # T is the maturity time
    # K is the strike price of the put option
    # stockprice is 2 dimensional numpy array containing all possible stock values throughout the time steps up until time T

    q = (1+r1-d)/(u-d) #risk neutral probability for up

    stockprice = np.full((T+1,T+1),np.nan)
    stockprice[0,0] = S_0
    for j in range(1,T+1):
        for i in range(0,j):
            stockprice[i,j] = stockprice[i,j-1]*u
            stockprice[i+1,j] = stockprice[i,j-1]*d
    # print("Stock prices:")
    # print(np.around(stockprice, 2)) #just for visualisation

    putoption = np.full((T+1,T+1),np.nan)
    putoption[:,-1] = np.where(K-stockprice[:,-1]<0,0,K-stockprice[:,-1]) # V_0
    for j in range(T-1,-1,-1):
        for i in range(0,j+1):
            putoption[i,j] = (q*putoption[i,j+1]+(1-q)*putoption[i+1,j+1])/(1+r1) # Explicit expected value
            # print(i, j)
            # print(putoption[i,j])
    # print("Put option prices:")
    # print(np.around(putoption, 2)) #just for visualisation
    return stockprice, putoption
```

```python
# Parameters

S_0 = 1000
r1 = 0.02/365
T = 7
K = 990

d_values = [0.95, 0.99]
u_values = [1.02, 1.03, 1.05, 1.101]
```

```python
import matplotlib.pyplot as plt

results_put = {}

for d in d_values:
    put_prices = []
    for u in u_values:
        stockprice, putoption = put_option(S_0, r1, u, d, T, K)
        put_prices.append(putoption[0, 0])  # fair price at t=0
    results_put[d] = put_prices

# Plotting
plt.figure(figsize=(8, 5))
for d in d_values:
    plt.plot(u_values, results_put[d], marker='o', label=f'd = {d}')

plt.title('European Put Option Value vs u for different d values')
plt.xlabel('u (Up factor)')
plt.ylabel('Put Option Price V₀')
plt.legend()
plt.grid(True)
plt.show()
```

## (b)

- Increasing $u$ can make the put option more valuable because, under the risk-neutral measure, a larger $u$ decreases the probability of an up-move ($q$) and increases the probability of a down-move. More weight on down-moves increases the expected payoff of the put.

- Decreasing $d$ still makes the put option more valuable because larger down-moves decrease possible stock prices and increase the chance that the option finishes in the money.

In summary, a higher $u$ can increase the put price, and a lower $d$ increases the put price.

```python
def call_option(S_0, r1, u, d, T, K):
    # This function simulates all possible stock trajectories in a T-period binomal model and
    # determines the corresponding European call option fair prices corresponding to the payoffs

    # S_0 is the initial stock price
    # r_1 is the daily interest rate used as 1+r1 multiplier
    # u is the amount by which the stock goes up
    # d is the amount by which the stock goes down
    # T is the maturity time
    # K is the strike price of the call option
    # stockprice is 2 dimensional numpy array containing all possible stock values throughout the time steps up until time T

    q = (1+r1-d)/(u-d) #risk neutral probability for up

    stockprice = np.full((T+1,T+1),np.nan)
    stockprice[0,0] = S_0
    for j in range(1,T+1):
        for i in range(0,j):
            stockprice[i,j] = stockprice[i,j-1]*u
            stockprice[i+1,j] = stockprice[i,j-1]*d
    # print("Stock prices:")
    # print(np.around(stockprice, 2)) #just for visualisation

    calloption = np.full((T+1,T+1),np.nan)
    calloption[:,-1] = np.where(stockprice[:,-1]-K<0,0,stockprice[:,-1] -K) # V_0
    for j in range(T-1,-1,-1):
        for i in range(0,j+1):
            calloption[i,j] = (q*calloption[i,j+1]+(1-q)*calloption[i+1,j+1])/(1+r1) # Explicit expected value
            # print(i, j)
            # print(calloption[i,j])
    # print("Call option prices:")
    # print(np.around(calloption, 2)) #just for visualisation
    return stockprice, calloption
```

```python
results_call = {}

for d in d_values:
    call_prices = []
    for u in u_values:
        stockprice, calloption = call_option(S_0, r1, u, d, T, K)
        call_prices.append(calloption[0, 0])  # fair price at t=0
    results_call[d] = call_prices

# Plotting
plt.figure(figsize=(8, 5))
for d in d_values:
    plt.plot(u_values, results_call[d], marker='o', label=f'd = {d}')

plt.title('European call Option Value vs u for different d values')
plt.xlabel('u (Up factor)')
plt.ylabel('call Option Price V₀')
plt.legend()
plt.grid(True)
plt.show()
```

- Increasing $u$ makes the call option more valuable because a larger $u$ raises the possible stock prices in the up-moves. Even though the probability of an up-move ($q$) decreases slightly, the higher stock prices on the up paths dominate, increasing the expected payoff of the call.

- Decreasing $d$ can make the call option more valuable because, under the risk-neutral measure, a smaller $d$ increases the probability of an up-move ($q$) and decreases the probability of a down-move. More weight on up-moves increases the expected payoff of the call.

In summary, a higher $u$ increases the call price, and a smaller $d$ also increases the call price.
