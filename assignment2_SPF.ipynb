The Langevin SDE is:

$$dX_t = c X_t dt + \sigma dW_t.$$

We want to check that the following process solves it:

$$ X_t = X_0 e^{ct} + \sigma e^{ct} \int_0^t e^{-cs}\,dWs.$$


We check for the initial conditon $(X(t = 0))$ and apply the It√¥ Formula to the given process to show that it solves the SDE for a constant initial condition $X_0$:

$$ X_0 = X_0e^{c\cdot 0} + \sigma e^{c \cdot 0} \int_0^0 e^{-c\cdot s}\,dW_s = X_0.$$
\begin{align*}
	dX_t &= \left[c\cdot X_0 e^{ct} + \sigma c e^{ct} \int_0^t e^{-cs}\, dW_s\right]dt + \sigma e^{ct} e^{-ct} dW_t + \frac{1}{2}\cdot 0\\& = c \left[X_0e^{ct} + \sigma e^{ct} \int_0^t e^{-cs}dW_s\right]dt + \sigma dW_t = cX_t dt + \sigma dW_t.
\end{align*}

Thus, the process solves the Langevin SDE.

The Vasicek interest rate model SDE is:

$$dr_t = c (\mu - r_t)dt + \sigma dW_t.$$

We want to check that the following process solves it:

$$ r_t = r_0 e^{-ct} + \mu (1-e^{-ct})+ \sigma e^{-ct}\int_0^t e^{cs}dW_s.$$


We check for the initial conditon $(r(t = 0))$ and apply the It\^{o} Formula to the given process to show that it solves the SDE for a constant initial condition $r_0$:

$$ r_0 = r_0e^{-c\cdot 0} + \mu (1-e^{-c\cdot 0}) + \sigma e^{-c\cdot 0} \int_0^0 e^{cs}\,dW_s = r_0.$$

\end{align*}

Thus, the process solves the Vasicek SDE.

We want to determine the fair price a binary spread with payoff function $C(S_T) = K \mathbb{1}_{(a,b)}(S_T)$ in the Black-Scholes model setting. We will use the fact that $$\frac{S_T}{S_t} = e^{\left(r-\frac{1}{2}\sigma^2\right)(T - t) + \sigma \sqrt{T-t}Z}$$ where $Z \sim N(0,1)$ is independent of the filtration at time $t$, $\mathcal{F}_t$ under the risk neutral measure $\mathbb{Q}$.


\begin{align*}
V_t &= \mathbb{E}_{\mathbb{Q}}\left[e^{-r(T-t)} C(S_T)|\mathcal{F}_t\right] = \mathbb{E}_{\mathbb{Q}}\left[e^{-r(T-t)} K \mathbb{1}_{(a,b)}(S_T)|\mathcal{F}_t\right] = e^{-r(T-t)} K\mathbb{E}_{\mathbb{Q}}\left[ \mathbb{1}_{(a,b)}(S_T)|\mathcal{F}_t\right]
\end{align*}

We compute the expectation term 
$\mathbb{E}_{\mathbb{Q}}\left[ \mathbb{1}_{(a,b)}(S_T)\mid\mathcal{F}_t\right]$ 
separately. By independence of the ratio $R = \tfrac{S_T}{S_t}$ from $\mathcal{F}_t$ 
and the Markov property of $S_t$, the conditional law of $S_T$ given 
$\mathcal{F}_t$ depends only on $S_t$. Defining $
g(x) := \mathbb{E}_{\mathbb{Q}}\!\left[\mathbb{1}_{(a,b)}(xR)\right]$,
we obtain
$$
\mathbb{E}_{\mathbb{Q}}\!\left[\mathbb{1}_{(a,b)}(S_T)\mid\mathcal{F}_t\right]
= g(S_t)
= \mathbb{E}_{\mathbb{Q}}\!\left[
\mathbb{1}_{\left(a < x \cdot \tfrac{S_T}{S_t} < b\right)}
\right]_{x = S_t}.
$$

So

\begin{align*}
    &\mathbb{E}_{\mathbb{Q}}\left[ \mathbb{1}_{(a,b)}(S_T)|\mathcal{F}_t\right] = \mathbb{E}_{\mathbb{Q}}\left[ \mathbb{1}_{(a < S_T < b)}|\mathcal{F}_t\right] = \mathbb{E}_{\mathbb{Q}}\left[ \mathbb{1}_{\left(a < S_T\cdot \frac{S_t}{S_t} < b\right)}\bigg|\mathcal{F}_t\right] = \mathbb{E}_{\mathbb{Q}}\left[ \mathbb{1}_{\left(a < x \cdot \frac{S_T}{S_t} < b\right)}\right] \bigg|_{x = S_t}\\
    &\quad = \mathbb{Q}\left(a < x \cdot \frac{S_T}{S_t} < b \right) = \mathbb{Q}\left(\frac{a}{x} < \frac{S_T}{S_t} < \frac{b}{x} \right)  = \mathbb{Q}\left(\frac{a}{x} < e^{\left(r-\frac{1}{2}\sigma^2\right)(T - t) + \sigma \sqrt{T-t}Z} < \frac{b}{x} \right)\\
    &\quad = \mathbb{Q}\left(\frac{a}{x} e^{-\left(r-\frac{1}{2}\sigma^2\right)(T - t) + \sigma \sqrt{T-t}Z} < e^{\sigma \sqrt{T-t}Z} < \frac{b}{x} e^{-\left(r-\frac{1}{2}\sigma^2\right)(T - t) + \sigma \sqrt{T-t}Z} \right)\\
    &\quad = \mathbb{Q}\left(-\left(r-\frac{1}{2}\sigma^2\right) (T-t) \log \left(\frac{a}{x}\right) < \sigma \sqrt{T-t} Z < -\left(r-\frac{1}{2}\sigma^2\right) (T-t) \log \left(\frac{b}{x}\right)\right)\\
    &\quad = \mathbb{Q}\left(\frac{1}{\sigma}\left(r-\frac{1}{2}\sigma^2\right) \sqrt{T-t} \log \left(\frac{x}{a}\right) < Z < \frac{1}{\sigma}\left(r-\frac{1}{2}\sigma^2\right) \sqrt{T-t} \log \left(\frac{x}{b}\right)\right)\\
    &\quad = \mathbb{Q}\left(\alpha < Z < \beta\right) = \Phi\left( \beta \right) - \Phi\left(\alpha\right).\\
\end{align*}


Now replacing back $x$ for $S_t$, which is what we conditioned the expectation to, and getting the previous terms we get that

$$V_t = e^{-r(T-t)} K\mathbb{E}_{\mathbb{Q}}\left[ \mathbb{1}_{(a,b)}(S_T)|\mathcal{F}_t\right] =  e^{-r(T-t)} K \left(\Phi\left( \beta \right) - \Phi\left(\alpha\right)\right)$$

with $$\alpha = \frac{1}{\sigma}\left(r-\frac{1}{2}\sigma^2\right) \sqrt{T-t} \log \left(\frac{S_t}{a}\right)$$ and $$\beta = \frac{1}{\sigma}\left(r-\frac{1}{2}\sigma^2\right) \sqrt{T-t} \log \left(\frac{S_t}{b}\right).$$



$\textbf{Exercise 3.}$

By the Feynman--Kac theorem, given an SDE of the type
$$
dX_t = \beta(t, X_t)\,dt + \gamma(t,X_t)\,dW_t, \quad t \in [0,T],
$$
and a Borel measurable function $h(x)$ such that $\mathbb{E}[|h(X_T)|] < \infty$,
there exists a function $g(t,x):\mathbb{R}^2 \to \mathbb{R}$ such that
$$
g(t,X_t) = \mathbb{E}[h(X_T)\mid \mathcal{F}_t],
$$
and $g(t,x)$ satisfies the PDE
$$
g_t(t,x) + \beta(t,x)g_x(t,x) + \frac{1}{2}\gamma(t,x)^2 g_{xx}(t,x) = 0,
\qquad g(T,x) = h(x).
$$

In this exercise, we have the PDE
$$
f_t(t,x) + 2t\, f_x(t,x) + \frac{1}{2}t^4 f_{xx}(t,x) = 0,
\qquad f(T,x) = x^2.
$$
We identify
$$
\beta(t,x) = 2t, \qquad \frac{1}{2}\gamma(t,x)^2 = \frac{1}{2}t^4
\quad\Rightarrow\quad \gamma(t,x)=t^2,
$$
leading to the SDE
$$
dX_t = 2t\,dt + t^2\, dW_t.
$$
In integral form,
$$
X_T = X_t + (T^2 - t^2) + \int_t^T s^2\, dW_s,
$$

Thus by the Feynman--Kac representation, $f(t,x) = \mathbb{E}\big[X_T^2 \mid \mathcal{F}_t\big]$,

\begin{align*}
    &\mathbb{E}\left[X_T^2 |\mathcal{F}_t\right] = \mathbb{E}\left[\left(X_t + (T^2 - t^2) + \int_t^T s^2\,dW_s\right)^2 \bigg|\mathcal{F}_t\right]\\
    &\quad = \mathbb{E}\left[X_t^2+(T^2-t^2)^2 + 2X_t(T^2-t^2) + \left(\int_t^T s^2\,dW_s\right)^2 + 2X_t (T^2 - t^2) \int_t^T s^2\,dWs\bigg |\mathcal{F}_t\right]\\
    &\quad = X_t^2+(T^2-t^2)^2 + 2X_t(T^2-t^2) + \mathbb{E}\left[ \left(\int_t^T s^2\,dW_s\right)^2\bigg|\mathcal{F}_t\right]  + 0\\
    &\quad = X_t^2+(T^2-t^2)^2 + 2X_t(T^2-t^2) + \mathbb{E}\left[ \left(\int_t^T s^4\,ds\right)\bigg|\mathcal{F}_t\right]  \\
     &\quad = X_t^2+(T^2-t^2)^2 + 2X_t(T^2-t^2) + \frac{T^5-t^5}{5} = f(t,x).
\end{align*}


Thus
$$
f(t,x) = x^2 + 2x(T^2 - t^2) + (T^2 - t^2)^2 + \frac{T^5 - t^5}{5}.
$$

We verify the boundary condition:

$$f(T,x) = x^2 + (T^2 - T^2)^2 + 2x (T^2 - T^2) + \frac{T^5 - T^5}{5} = x^2.$$


Finally, we check that $f$ solves the PDE. We compute
$$
f_x = 2x + 2(T^2 - t^2), \qquad f_{xx} = 2,
$$
$$
f_t = -4t(T^2 - t^2) - 4tx - t^4.
$$
Then
\begin{align*}
f_t + 2t f_x + \frac{1}{2}t^4 f_{xx}
&= (-4t(T^2 - t^2) - 4tx - t^4)
   + 2t(2x + 2(T^2 - t^2)) + t^4 \\
&= 0.
\end{align*}

$$F_t(t,x) + r x F_x(t,x) + \frac{\sigma^2}{2} x^2 F_{xx}(t,x) - r F(t,x) $$
\begin{align*}
&e^{-r(T-t)}\cdot (r) \left[(2r-\sigma^2) (T-t) + \log(S_t^2)\right] - e^{-r(T-t)}(2r-\sigma^2) + rS_t \left[2e^{-r(T-t)}\frac{1}{S_t}\right]\\ &\quad  + \frac{\sigma^2}{2}S_t^2 \left[- 2e^{-r(T-t)}\frac{1}{S_t^2}\right] - r \left[e^{-r(T-t)}\,\left[ (2r-\sigma^2)(T-t) + \log\left(S_t^2\right) \right]\right]=0
\end{align*}

\begin{align*}
& r \left[(2r-\sigma^2) (T-t) + \log(S_t^2)\right] - (2r-\sigma^2) + 2r-\sigma^2 - r \left[ (2r-\sigma^2)(T-t) + \log\left(S_t^2\right) \right] = 0.
\end{align*}

where all terms cancel out.

Finally we check the boundary condition $F(T,x) = f(x)$ where $f(x)$ is the payoff function.

$$F(T,x) = e^{-r(T-T)}\,\left[ (2r-\sigma^2)(T-T) + \log\left(S_t^2\right) \right] = 1 \cdot \left[0 + \log(S_t^2)\right] = \log(S_t^2) = f(x).$$


So we have verified that our fair price solution satisfies the PDE boundary problem.

- *(a)*
We know that under $P$, $S_t$ satisfies:
$$
S_t = S_0\,e^{\mu t + \sigma W_t},
$$
and under $Q$:
$$
S_t = S_0\,e^{\left( r-\frac{\sigma^2}{2}\right) t + \sigma \tilde{W}_t},
$$
where both $W_t$ and $\tilde{W}_t$ are brownian motions under their respective measures.

We will use the following equations to simulate the increments of the geometric brownian motion:

Under $P$:
$$
\frac{S_t}{S_{t-\Delta t}} = \frac{S_0\,e^{\mu t + \sigma W_t}}{S_0\,e^{\mu (t-\Delta t) + \sigma W_{t-\Delta t}}} = e^{\mu \,\Delta t + \sigma (W_t-W_{t-\Delta t})} \qquad \forall t > 0, \; \Delta t \in (0,t).
$$

Under $Q$:
$$
\frac{S_t}{S_{t-\Delta t}} = \frac{S_0\,e^{\left( r-\frac{\sigma^2}{2}\right) t + \sigma \tilde{W}_t}}{S_0\,e^{\left( r-\frac{\sigma^2}{2}\right) (t-\Delta t) + \sigma \tilde{W}_{t-\Delta t}}} = e^{\left( r-\frac{\sigma^2}{2}\right) \,\Delta t + \sigma (\tilde{W}_t-\tilde{W}_{t-\Delta t})} \qquad \forall t > 0, \; \Delta t \in (0,t).
$$

Additionally, we will use that $\tilde{W}_t-\tilde{W}_{t-\Delta t}$, $W_t-W_{t-\Delta t} \sim \mathcal{N}(0, \Delta t)$.

```python
def GBM(s0 = 270, T = 30, steps = 30, mu = 0, sigma = 0.1, N = 100):
    rng = np.random.default_rng(42)
    dt = T/steps
    dB = rng.normal(0, np.sqrt(dt), (N,steps))
    stock_increments = np.exp(mu*dt + sigma*dB)

    initial_prices = np.ones([N,1])*s0
    paths = np.concatenate([initial_prices,stock_increments], axis =1)
    paths = np.cumprod(paths, axis = 1)
    return(paths)
```

We will simulate all $30+1$ time steps for the next section but only plot the first $25+1$, as required.

In contrast, since the drift component of $S_t$ under $Q$ is exactly $r - \sigma^2/2$, the discounting is mostly canceled out by the drift and the path stays, on average, just below the starting price $S_0$.

$$
d_1(\tau,x) = \frac{log(x/K) + (r+\sigma^2/2)\tau}{\sigma \sqrt{\tau}}, \qquad d_2(\tau,x) = d_1(\tau,x) - \sigma \sqrt{\tau}.
$$
Therefore, we compute the theoretical value using B-S and obtain $V_0 ‚âà 155.72$.

```python
from scipy.stats import norm
def european_claim(t = 0, T =30, K = 333, S = 270, sigma = 0.1):
    d1 = (np.log(S/K)+(r+sigma**2/2)*(T-t))/(sigma*np.sqrt(T-t))
    d2 = d1 - sigma*np.sqrt(T-t)
    return(S*norm.cdf(d1)-K*np.exp(-r*(T-t))*norm.cdf(d2))

european_claim()
```

We will again follow the scheme provided in 6.6 of the lecture notes. We will estimate the value of an european max-call option and provide a 90%-confidence interval on said estimate.

By the CLT and the LLN we have that $\sqrt{n}(\bar{c}_n - E_Q[C])/s_n\rightsquigarrow \mathcal{N}(0,1)$ as $n \to \infty$, where $s_n^2 = \frac{1}{n-1}\sum_{i=1}^n(C_i - \bar{c}_n)^2$.

Therefore, for $n$ large enough:
$$
Q\left(\sqrt{n}\lvert\frac{\bar{c}_n - E_Q[C]}{s_n}\rvert > 1.645\right) \approx 0.10.
$$
It follows that the following is an approximate 90%-CI for the price of the claim C:
$$
\left[e^{-rT}(\bar{c}_n - 1.645s_n/\sqrt{n}), \, e^{-rT}(\bar{c}_n + 1.645s_n/\sqrt{n})\right].
$$


We will assume that the base parameters are:

*   $S_0 = 270$
*   $r = 0.0349$
*   $T = 30$
*   $N_{steps} = 30+1$ (time steps)
*   $\mu = 0$
*   $\sigma = 0.1$
*   $K = 333$


```python
def european_max_call(prices, strike=333):
    max_path = np.max(prices, axis=0)
    return(np.maximum(0, max_path-strike))
```

```python
T = 30
steps = 30
sigma = 0.1
mu = 0
r = 0.0349
paths = 10000
K = 333


q95 = norm.ppf(0.95)
time = np.linspace(0, T, steps+1)

discount = np.exp(-r * time)
discount = np.repeat(discount[:, np.newaxis], paths, axis=1)

paths_Q = np.transpose(GBM(s0 = 270, T = T, steps = steps, mu = r-sigma**2/2, sigma = sigma, N = paths))
path_claim = european_max_call(paths_Q[1:,:], K)

mean_paths = np.mean(path_claim)
std_paths = np.std(path_claim, ddof=1)

mean_paths_disc = mean_paths*np.exp(-r*T)

print(f"The estimate of the value of the European max call is: {mean_paths_disc:.2f}")
left_CI = np.exp(-r*T)*(mean_paths - q95*std_paths/np.sqrt(paths))
right_CI = np.exp(-r*T)*(mean_paths + q95*std_paths/np.sqrt(paths))
print(f"The 90% CI for the estimate is: ({left_CI:.2f}, {right_CI:.2f})")
```

```python
### Volatility
T = 30
steps = 30
r = 0.0349

sigma = np.linspace(0, 4, steps*3)
mu = 0
paths = 100000
K = 333

estimates = np.empty(len(sigma))

for i in range(len(sigma)):
    paths_Q = np.transpose(GBM(s0 = 270, T = T, steps = steps, mu = r -sigma[i]**2/2, sigma = sigma[i], N = paths))
    path_claim = european_max_call(paths_Q[1:,:], K)
    mean_paths = np.mean(path_claim)
    mean_paths_disc = mean_paths*np.exp(-r*T)
    estimates[i] = mean_paths_disc

plt.figure(figsize=(8,5))
plt.plot(sigma, estimates, marker='o', linestyle='-', color='b', ms =1)
plt.vlines(1,min(estimates),max(estimates))
plt.title('European Max Call Estimate vs Sigma')
plt.xlabel(r'$\sigma$')
plt.ylabel('Estimated Max Call Value')
plt.grid(True)
plt.show()
```

Observe that the drift component $r-\sigma^2/2$ decreases quadratically as sigma becomes bigger, obviously this results in lower expected stock prices. Conversely, the volatility component is exactly $\sigma$, so it increases linearly. Moreover, as the volatility increases, the probability to achieve a greater maximum increases, so the fair value of the option rises.

In the plot we can see that for small values of $\sigma$, the fair value of the option increases with $\sigma$. In contrast, it decreases with $\sigma$ for larger values. This change in behavior happens around $\sigma \approx 1$, which is the point at which the drift component starts to decrease more rapidly than the volatility increases.

In conclusion, for small values of $\sigma$, the volatility compensates decreases in drift, but for larger values of $\sigma$ this is reverted.
